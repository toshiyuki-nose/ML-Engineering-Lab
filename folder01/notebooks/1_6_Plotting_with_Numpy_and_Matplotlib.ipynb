## 1_6_Plotting_with_Numpy_and_Matplotlib

```python
## Import NumPy and Matplotlib
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
## 1. The main matplotlib command can be represented in pseudocode as follows:
## plt.plot(NumPy array, NumPy array of the same length)
```

```python
## 2. Plot a straight line by providing two NumPy arrays of the same length
plt.plot(np.arange(10), np.arange(10))
```

```python
## 3. Plot the exponential function
plt.plot(np.arange(10), np.exp(np.arange(10)))
```

```python
## 4. Arrange two graphs side-by-side
plt.figure()
plt.subplot(121)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(122)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
## Or arrange them vertically
plt.figure()
plt.subplot(211)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(212)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

Here, after instantiating the figure using the `plt.figure` function, the grid size is specified using the first two numbers of the `subplot` command.
The grid size indicated by `plt.subplot(221)` is 2x2 (based on the first two numbers).
The final number specifies which grid position to plot in (ordered from left to right and top to bottom).

```python
## 5. Plot on a 2x2 grid (positions 1–4) in order from left to right and top to bottom.

plt.figure()
plt.subplot(221)
plt.plot(np.arange(10), np.exp(np.arange(10)))
plt.subplot(222)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(223)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
plt.subplot(224)
plt.scatter(np.arange(10), np.exp(np.arange(10)))
```

```python
## 6. Finally, let's try it with actual data.
from sklearn.datasets import load_iris
iris = load_iris()
data = iris.data
target = iris.target
# Resize the figure for better viewing
plt.figure(figsize=(12,5))
# First subplot
plt.subplot(121)
# Visualize the first two columns of data:
plt.scatter(data[:,0], data[:,1], c=target)
# Second subplot
plt.subplot(122)
# Visualize the last two columns of data:
plt.scatter(data[:,2], data[:,3], c=target)
```

For the `c` parameter, specify an array of colors.
In this case, specify the target consisting of the three numbers 0, 1, and 2 from the iris dataset.
