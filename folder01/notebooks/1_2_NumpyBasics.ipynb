## 1_2_NumpyBasics

### Numpy array shape and dimensions

```python
## 1. Importing the numpy library
import numpy as np
```

```python
## 2. This generates a Numpy array containing 10 numbers. It's equivalent to Python's range(10) method.
np.arange(10)
```

```python
## 3. Store this array in a variable and view its shape using `shape`.
array_1 = np.arange(10)
array_1.shape
```

```python
## 4. Get the number of dimensions of array_1
array_1.ndim
```

```python
## 5. This array contains 10 elements.
## Let's try changing the shape of the array by calling the reshape method.
array_1.reshape((5,2))
```

```python
## 6. The array's structure has been changed to a 5x2 data object.
## Save this modified array.
array_1 = array_1.reshape((5,2))
```

```python

## 7. Let's re-examine the dimensions of array_1.
## Note that it's now 2 dimensions.
array_1.ndim 
```

### NumPy broadcasting

```python
## 8. Use broadcasting to add 1 to each element of each array.
## Note that changes to the array will not be saved.
array_1 + 1
```

```python
## 9. Create a new array called array_2.
## Multiply this array by itself.
## All elements have been squared.

array_2 = np.arange(10)
array_2 * array_2
```

```python
## 10. Note that this is equivalent to array_2 * array_2
array_2 = array_2 ** 2  
array_2 = array_2.reshape((5,2))
array_2
```

```python
## 11. Let's try changing array_1 as well.
array_1 = array_1 + 1
array_1
```

```python
## 12. Let's try adding array_1 and array_2 elements together.
array_1 + array_2
```

### NumPy initialization and dtype

```python
## 14. Use np.zeros to initialize the array with zeros.
np.zeros((5,2))
```

```python
## 15. Use `np.ones` to initialize an array with 1s.
## Adding the `dtype` parameter and setting it to `np.int64` makes the value 1 a NumPy integer type.
## `dtype` represents the type of all elements in a NumPy array.
## The elements of an array remain the same while the array is being used.
np.ones((5,2), dtype = np.int64)
```

```python
## 16. Use `np.empty` to allocate memory for an array of a specified size and dtype, 
## but without initializing the elements with specific values.
np.empty((5,2), dtype = float)
```

### INDEX

```python
## 17. Access the element at the first row and first column of array_1.
array_1[0,0]    #Finds value in first row and first column.
```

```python
## 18. Display the first line.
array_1[0,:]
```

```python
## 19. Display the first column.
array_1[:,0]
```

```python
## 20. Display the values ​​for both axes (rows and columns) from row 3 to row 5.
array_1[2:5, :]
```

```python
## 21. Display only the values ​​in the first column from rows 3 to 5.
array_1[2:5,0]
```

### Boolean array
Furthermore, NumPy handles indices as Boolean values ​​(logical values).

```python
## 22. First, create a Boolean array from array_1.
array_1 > 5
```

```python
## 23. To filter using an array of Boolean values, enclose the array of Boolean values ​​in square brackets ([]).
array_1[array_1 > 5]
```

### Arithmetic operations

```python
## 24. Use the sum method to sum up all elements of array_1.
array_1.sum()
```

```python
## 25. Calculate the sum for each row.
array_1.sum(axis = 1)
```

```python
## 26. Calculate the sum for each column.
array_1.sum(axis = 0)
```

```python
## 27. Similarly, we find the mean of each column. Note that the dtype of the array mean is np.float.
array_1.mean(axis = 0)
```

### NaN値

```python
## 28. scikit-learn does not accept np.nan values. Therefore, we create array_3 as follows:
array_3 = np.array([np.nan, 0, 1, 2, np.nan])
```

```python
## 29. To search for NaN values, use a special Boolean array created by the np.isnan function
##  (True represents NaN values).
np.isnan(array_3)
```

```python
## 30. To exclude NaN values, use the tilde symbol (~) to perform a logical negation of the elements in this Boolean array. 
## When doing so, enclose the negation expression in square brackets ([]).
array_3[~np.isnan(array_3)]
```

```python
## 31. Alternatively, set the NaN value to 0.
array_3[np.isnan(array_3)] = 0
array_3
```
