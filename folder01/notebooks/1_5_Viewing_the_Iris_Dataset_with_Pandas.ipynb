## 1_5_Viewing_the_Iris_Dataset_with_Pandas

```python
import numpy as np    #Load the numpy library for fast array computations
import pandas as pd   #Load the pandas data-analysis library
import matplotlib.pyplot as plt   #Load the pyplot visualization library

%matplotlib inline

from sklearn import datasets
iris = datasets.load_iris()
```

```python
## 1. Create a data frame using the explanatory variable `iris.data`. When creating the data frame, specify the column names using the `columns` argument.

iris_df = pd.DataFrame(iris.data, columns = iris.feature_names)
```

```python
## 2. Let's create a histogram of the 'sepal length' values ​​from the data frame.

iris_df['sepal length (cm)'].hist(bins=30)
```

```python
## 3. You can also color-code this histogram based on the target.

for class_number in np.unique(iris.target):
    plt.figure(1)
    iris_df['sepal length (cm)'].iloc[np.where(iris.target == class_number)[0]].hist(bins=30)
```
