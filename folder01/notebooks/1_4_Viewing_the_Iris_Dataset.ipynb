```python
import numpy as np    #Load the numpy library for fast array computations
import pandas as pd   #Load the pandas data-analysis library
import matplotlib.pyplot as plt   #Load the pyplot visualization library

%matplotlib inline

from sklearn import datasets
iris = datasets.load_iris()
```

```python
## 1. To access the explanatory variables (observed variables), enter the following command.
## This will output a NumPy array.

iris.data
```

```python
## 2. Let's examine this NumPy array.
iris.data.shape
```

In other words, the shape of this data is 150 rows by 4 columns.

```python
## 3. Let's examine the first row of this data.
iris.data[0]
```

The first row of this NumPy array consists of four numbers.

```python
## 4. Enter the following command to identify their meanings.
iris.feature_names
```

```python
## 5. Next, let's examine the target variable (output variable) in the same way.
iris.target
```

An array consisting of the three numbers 0, 1, and 2 is output.

```python
## 6. Executing the following command reveals that the shape of the target variable is (150, ).
iris.target.shape
```

```python
## 7. Investigate what the numbers in this array represent.
iris.target_names
```
