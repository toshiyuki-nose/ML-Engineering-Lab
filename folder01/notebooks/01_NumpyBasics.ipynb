## 01_NumpyBasics

### NumpPy配列の形状と次元

```python
## Importing the numpy library
import numpy as np
```

```python
## This generates a Numpy array containing 10 numbers. It's equivalent to Python's range(10) method.
np.arange(10)
```

```python
# Store this array in a variable and view its shape using `shape`.
array_1 = np.arange(10)
array_1.shape
```

```python
#Get the number of dimensions of array_1
array_1.ndim
```

```python
# This array contains 10 elements.
# Let's try changing the shape of the array by calling the reshape method.
array_1.reshape((5,2))
```

```python
# The array's structure has been changed to a 5x2 data object.
# Save this modified array.
array_1 = array_1.reshape((5,2))
```

```python

# Let's re-examine the dimensions of array_1.
# Note that it's now 2 dimensions.
array_1.ndim 
```

### NumPyのブロードキャスト

```python
# Use broadcasting to add 1 to each element of each array.
# Note that changes to the array will not be saved.
array_1 + 1
```

```python
# Create a new array called array_2.
# Multiply this array by itself.
# All elements have been squared.

array_2 = np.arange(10)
array_2 * array_2
```

```python
# Note that this is equivalent to array_2 * array_2
array_2 = array_2 ** 2  
array_2 = array_2.reshape((5,2))
array_2
```

```python
# Let's try changing array_1 as well.
array_1 = array_1 + 1
array_1
```

```python
# Let's try adding array_1 and array_2 elements together.
array_1 + array_2
```

### NuPyの初期化とdtype

```python
# Use np.zeros to initialize the array with zeros.
np.zeros((5,2))
```

```python
# Use `np.ones` to initialize an array with 1s.
# Adding the `dtype` parameter and setting it to `np.int64` makes the value 1 a NumPy integer type.
# `dtype` represents the type of all elements in a NumPy array.
# The elements of an array remain the same while the array is being used.
np.ones((5,2), dtype = np.int64)
```

```python
# Use `np.empty` to allocate memory for an array of a specified size and dtype, 
# but without initializing the elements with specific values.
np.empty((5,2), dtype = float)
```
